# DIM DATE

```python
orders.head()
```

```python
order_date = orders.copy()
```

```python
# Lấy date trong purchase date
order_date["full_date"] = order_date["order_purchase_timestamp"].dt.strftime("%Y-%m-%d")
```

```python
# Tạo ra id từ cột purchase date 
order_date["date_id"] = order_date["order_purchase_timestamp"].dt.strftime("%Y%m%d")
```

```python
# Tạo ra năm, tháng, quý, ngày
order_date["year"] = order_date["order_purchase_timestamp"].dt.year
order_date["month"] = order_date["order_purchase_timestamp"].dt.month
order_date["quarter"] = order_date["month"].apply(lambda x : 1 if x in [1,2,3] 
                                                        else 2 if x in [4,5,6]
                                                        else 3 if x in [7,8,9]
                                                        else 4)
order_date["day"] = order_date["order_purchase_timestamp"].dt.day
```

```python
# Lấy tuần 
order_date["week"] = (order_date["day"]/7 + 1).astype(int)
```

```python
# Lấy thứ
order_date["weekday"] = order_date["order_purchase_timestamp"].dt.day_name()
```

```python
month_map = {
    1: "January",
    2: "February",
    3: "March",
    4: "April",
    5: "May",
    6: "June",
    7: "July",
    8: "August",
    9: "September",
    10: "October",
    11: "November",
    12: "December"
}
```

```python
# Đổi tên tháng
order_date["month_name"] = order_date["month"].map(month_map)
```

```python
# Check có phải cuổi tuần không
order_date["is_weekend"] = order_date["weekday"].apply(lambda x : True if x in ["Saturday", "Sunday"] else False)
```

```python
order_date.head()
```

```python
# Chọn các cột cần thiết cho dim date
dim_date = order_date[[
                        "date_id",
                        "full_date",
                        "year",
                        "quarter",
                        "month",
                        "month_name",
                        "week",
                        "day",
                        "weekday",
                        "is_weekend"
                        ]]
```

```python
dim_date.head()
```

```python
dim_date.to_csv("../data_warehouse/dim_date.csv", index = False, encoding = "utf-8")
```

# DIM SALES

```python
payments.head()
```

```python
# Tạo cột danh mục thanh toán
payments["payment_category"] = payments["payment_installments"].apply(lambda x : "Installment" if x > 1 
                                                                                else "One-time")
```

```python
payments.head()
```

```python
# Chọn ra cột cần thiết cho dim_payment
dim_payment = payments[["payment_type",	"payment_installments",	"payment_category"]].drop_duplicates().reset_index(drop = True)
```

```python
# Tạo cột id
dim_payment["payment_id"] = dim_payment.index + 1
```

```python
dim_payment = dim_payment[["payment_id", "payment_type",	"payment_installments",	"payment_category"]]
```

```python
dim_payment.head()
```

# FACT SALES

```python
order_items.head()
```

```python
order_items.info()
```

```python
# Join với bảng reviews
fact_join_review = pd.merge(order_items, reviews, on = "order_id", how = "left")
```

```python
# Join với bảng orders để lấy customer_id
fact_join_order = pd.merge(fact_join_review, orders, on = "order_id", how = "left")
```

```python
payments.head()
```

```python
# Join với bảng payments
fact_join_payment = pd.merge(fact_join_order, payments, on = "order_id", how = "left")
```

```python
# Join với bảng dim_payment để lấy payment_id
fact_join_payment = pd.merge(fact_join_payment, dim_payment, on =  ["payment_type",
                                                                    "payment_installments",	
                                                                    "payment_category"], how = "left")
```

```python
fact_join_payment.info()
```

```python
fact_join_payment["date_id"] = fact_join_payment["order_purchase_timestamp"].dt.strftime("%Y%m%d")
```

```python
fact_join_payment.info()
```

```python
# Tạo cột tổng tiền 
fact_join_payment["total_amount"] = fact_join_payment["price"] + fact_join_payment["freight_value"]
```

```python
# Tạo cột số ngày giao hàng
fact_join_payment["delivery_days"] = (fact_join_payment["order_delivered_customer_date"] - fact_join_payment["order_purchase_timestamp"]).dt.days
```

```python
# Tạo cột số ngày để tiếp nhận đơn hàng
fact_join_payment["approval_days"] = (fact_join_payment["order_approved_at"] - fact_join_payment["order_purchase_timestamp"]).dt.days
```

```python
# Tạo cột kiểm tra đơn giao muộn
fact_join_payment["is_late_delivery"] = (fact_join_payment["order_delivered_customer_date"] > fact_join_payment["order_estimated_delivery_date"])
```

```python
# Chọn các cột cần cho bảng fact sales
fact_sales = fact_join_payment[[
    "order_id",
    "customer_id",
    "product_id",
    "seller_id",
    "review_id",
    "payment_id",
    "date_id",
    "order_status",
    "price",
    "freight_value",
    "total_amount",
    "payment_value",
    "delivery_days",
    "approval_days",
    "is_late_delivery"
    ]]
```

```python
fact_sales.info()
```

```python
fact_sales = fact_sales.astype({"delivery_days":"Int64", "approval_days":"Int64"})
```
